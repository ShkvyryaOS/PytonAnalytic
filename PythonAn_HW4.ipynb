Условие 3: 3 задача Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.head()
```

```python
sns.barplot(x='view', y='price', data=df);
```

В зависимости от вида стоимость домов возрастает. значительно вышецена с оценкой вида 4. Виды с оценкой 1 и 2 не отличаются

```python
import numpy as np

corr_matrix = df.corr()

corr_matrix = np.round(corr_matrix, 1)

corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Наиболее выраженная прямая линейная зависимость стоимости от жилой площади, состояния конструкций дома и  жилой площади дома соседних домов, количества ванных комнат.В меньшей степени прямая зависимость прослеживается в зависимости от вида. 

```python
price = df['price']
sns.boxplot(x=df["grade"], y=price);
```

С улучшением качества конструкции и дизайна растет стоимость домов, разброс стоимости так же увеличивается. Медианная стоимость дома с отличным состояние конструкций примерно равна 3000000, бОльшая часть домов с таким состоянием конструкций имеет цену от 2500000 до 3500000 

```python
sns.jointplot(y=df['price'], x=df['sqft_living'], kind='reg');
```

Наблюдается прямая зависимость стоимости  с увеличением жилой площади. Основная масса домов имеет площадь до 4000 с ценой до 1000000

```python
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('waterfront')
plt.title('Зависимость стоимости домов от наличия вида на набережную');
```

Стоимость жилья с видом на набережную выше, чем без негою медианная  стоимость дома с видом на набережную примерно 1300000, ,  без вида около 600000
